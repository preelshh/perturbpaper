```python
df.rename(columns = {"Unnamed: 0" : "Cell_Type"}, inplace = True)
df = df.set_index("Cell_Type")
```

# CD 4 Activation TFs

```python
delta_cd4 = df.loc["CD4_act"] - df.loc["CD4_rest"]
```

```python
cd4_activation = delta_cd4.sort_values(ascending=False)
cd4_resting = delta_cd4.sort_values(ascending=True)

print("Top CD4 activation TFs:")
print(cd4_activation.head(20))

print("\nTop CD4 resting TFs:")
print(cd4_resting.head(20))
```

## Shared Activation TFs

```python
print(delta_cd4.describe())
print(delta_cd8.describe())
```

CD8 does not look very specific. CD4 Seems to have stronger TF rewiring. Most TFs decrease in specificty upon activation.A few TFs increase, and some TFs strongly decrease.

```python
shared = pd.DataFrame({
    "Δ_CD4": delta_cd4,
    "Δ_CD8": delta_cd8
})

shared = shared[(shared["Δ_CD4"] > 0.05) & (shared["Δ_CD8"] > 0.05)]
shared.sort_values("Δ_CD4", ascending=False).head(20)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

for ax, delta, label in zip(axes, [delta_cd4, delta_cd8], ['CD4', 'CD8']):
    values = delta.values
    kde = gaussian_kde(values)
    x = np.linspace(values.min() - 0.02, values.max() + 0.02, 500)

    ax.fill_between(x, kde(x), alpha=0.3, color='steelblue')
    ax.plot(x, kde(x), color='steelblue', lw=2)

    # mark the top/bottom 20 cutoffs
    sorted_vals = np.sort(values)
    act_cutoff = sorted_vals[-20]  # 20th highest
    rest_cutoff = sorted_vals[19]  # 20th lowest

    ax.axvline(act_cutoff, color='red', ls='--', lw=1.5,
               label=f'Top 20 act cutoff ({act_cutoff:.4f})')
    ax.axvline(rest_cutoff, color='blue', ls='--', lw=1.5,
               label=f'Top 20 rest cutoff ({rest_cutoff:.4f})')
    ax.axvline(0, color='gray', ls=':', lw=1)

    # rug plot of individual TFs
    ax.scatter(values, np.full_like(values, -0.5), marker='|',
               color='black', alpha=0.4, s=30)

    ax.set_title(f'{label} delta(RSS) Distribution (n={len(values)})')
    ax.set_xlabel('delta(RSS) = RSS_act - RSS_rest')
    ax.legend(fontsize=8)

axes[0].set_ylabel('Density')
plt.tight_layout()
plt.savefig('delta_rss_density.png', dpi=150, bbox_inches='tight')
plt.show()

# Print some stats to help pick a cutoff
for delta, label in zip([delta_cd4, delta_cd8], ['CD4', 'CD8']):
    values = delta.values
    mean = values.mean()
    std = values.std()
    print(f'{label}: mean={mean:.5f}, std={std:.5f}')
    print(f'  +1 SD cutoff: {mean + std:.5f}  ({(values > mean + std).sum()} TFs above)')
    print(f'  +2 SD cutoff: {mean + 2*std:.5f}  ({(values > mean + 2*std).sum()} TFs above)')
    print(f'  -1 SD cutoff: {mean - std:.5f}  ({(values < mean - std).sum()} TFs below)')
    print(f'  -2 SD cutoff: {mean - 2*std:.5f}  ({(values < mean - 2*std).sum()} TFs below)')
    print()
```
